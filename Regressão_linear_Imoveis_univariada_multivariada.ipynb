# Analisando variável por meio de testes de significância Teste F

O teste F é um teste de hipósete utilizado para comparação de variâncias. Serve para comparar duas estimativas de variâncias independentes

\begin{align}
F{} = \frac{MSR}{MSE}
\end{align}




referência1<https://www.fcav.unesp.br/Home/departamentos/cienciasexatas/AMANDALIZPACIFICOMANFRIM/aula-4.pdf>

referência2< https://edisciplinas.usp.br/pluginfile.php/2034075/mod_resource/content/1/An%C3%A1liseMultivariada-Aula06.pdf>

```python
X = imoveis[['Avg. Area Income', 'Avg. Area House Age', 'Avg. Area Number of Rooms',
       'Avg. Area Number of Bedrooms', 'Area Population']]
y=imoveis['Price']
```

```python
from sklearn.feature_selection import SelectPercentile
from sklearn.feature_selection import f_regression
Selector_f = SelectPercentile(f_regression, percentile=25)
Selector_f.fit(X,y)
for n,s,i in zip(imoveis.columns,Selector_f.scores_,Selector_f.pvalues_):
    print(n,s,n,i)

```

```python
pd.DataFrame( zip(imoveis.columns,Selector_f.scores_,Selector_f.pvalues_),
columns=['features', 'F-score','P-values'])
```

# Avaliando o primeiro modelo

Duas medidas de avaliação

Erro absoluto quadrático

Erro médio quadrático: é definido como sendo a média da diferença entre o valor do estimador e do parâmetro ao quadrado.
Assim, para um estimador, temos:
\begin{align}
MSE{} = E[(\Theta \hat{} - \Theta ) ^2 ]
\end{align}
<https://edisciplinas.usp.br/pluginfile.php/2250126/mod_resource/content/0/Aula20-2016-SLIDES.pdf>

Erro médio absoluto  MAD mede o afastamento médio das previsões em relação aos valores observados, constituindo na média dos erros da previsão. Neste sentido, seu valor ideal seria igual a zero. Assim, o erro médio tende a ser pequeno na medida em que os erros negativos e positivos se compensem. Por fim, conforme o mesmo autor, essa medida de acurácia pode ser
definida pela equação:
\begin{align}
MAE{} = \sum [A(t) - F(t)/n]
\end{align}

<https://www.ufjf.br/engenhariadeproducao/files/2014/09/2010_3_Dimas.pdf>


```python
from sklearn import metrics
print('MAE:%.3f'% metrics.mean_absolute_error(y_test, predictions,multioutput='raw_values'))
print('MSE:%.3f'% metrics.mean_squared_error(y_test, predictions,multioutput='raw_values'))
print('RMSE:%.3f'% np.sqrt(metrics.mean_squared_error(y_test, predictions,multioutput='raw_values')))
```

```python
MAE1 = (metrics.mean_absolute_error(y_test, predictions))
MSE1 = (metrics.mean_squared_error(y_test, predictions))
```

# Avaliando o quarto modelo
Duas medidas de avaliação



Erro absoluto quadrático

Erro médio quadrático: é definido como sendo a média da diferença entre o valor do estimador e do parâmetro ao quadrado.
Assim, para um estimador, temos:
\begin{align}
MSE{} = E[(\Theta \hat{} - \Theta ) ^2 ]
\end{align}
<https://edisciplinas.usp.br/pluginfile.php/2250126/mod_resource/content/0/Aula20-2016-SLIDES.pdf>

Erro médio absoluto  MAD mede o afastamento médio das previsões em relação aos valores observados, constituindo na média dos erros da previsão. Neste sentido, seu valor ideal seria igual a zero. Assim, o erro médio tende a ser pequeno na medida em que os erros negativos e positivos se compensem. Por fim, conforme o mesmo autor, essa medida de acurácia pode ser
definida pela equação:
\begin{align}
MAE{} = \sum [A(t) - F(t)/n]
\end{align}

<https://www.ufjf.br/engenhariadeproducao/files/2014/09/2010_3_Dimas.pdf>
